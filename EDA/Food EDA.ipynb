```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {'Model':['EfficientNetB2', 'ResNet50', 'MobileNetV2','Xception', 'VGG16'],
        'Accuracy':[0.53, 0.44, 0.12, 0.06,0.27],
        'Running Time':[2701,2759,2220,2534,2774]}
data = pd.DataFrame(data)
```

```python
# plt.figure(figsize=(12,12))

sns.set(rc={'figure.figsize':(10,5)}, font_scale=1.5)
sns.set_style({'axes.facecolor':'white', 'grid.color': '.8', 'font.family':'Times New Roman'})

p1 = sns.scatterplot('Running Time', # Horizontal axis
       'Accuracy', # Vertical axis
       data=data, # Data source
       size = 8,
       legend=False)

p1.set(title='Accuracy vs. Running Time of benchmark models')  

for line in range(0,data.shape[0]):
     p1.text(data['Running Time'][line]+0.04, data['Accuracy'][line], 
     data['Model'][line], horizontalalignment='left', 
     size='medium', color='#713c19', weight='semibold')
        
plt.savefig("Banchmark model.jpg", dpi=300, transparent=True)
```
